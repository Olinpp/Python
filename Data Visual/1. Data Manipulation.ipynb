```python
#menampilkan rata-rata quantity tansaksi berdasarkan brand
#menggunakan as_index=False agar memuat garis tabel
datame.groupby(['brand'], as_index=False)['quantity'].mean()
```

```python
#menampilkan banyak data yang tersedia berdasarkan brand
ps.sqldf("""
select brand, count(brand) from datame
group by brand
""")
```

### <font color='darkred'>5. Membuat kolom baru berdasarkan kondisi yang diinginkan </font>

#### Tambahkan kolom baru bernama desc_brand, berdasarkan brand, berdasrkan kondisi berikut:

|brand|desc_brand|
|---|---|
|BRAND_A|Sari Wangi|
|BRAND_B|Segar Sari|
|BRAND_C|Teh Sisri|
|BRAND_H|Nutri Sari|
|BRAND_J|Kacang Dua Kelinci|
|BRAND_L|Lays|
|BRAND_P|Japota|
|BRAND_R|Milkita|
|BRAND_S|Super Zuper|
|BRAND_W|Jagoan Neon|

```python
# Define Condition
condition = [
    (datame['brand'] == 'BRAND_A'),
    (datame['brand'] == 'BRAND_B'),
    (datame['brand'] == 'BRAND_C'),
    (datame['brand'] == 'BRAND_H'),
    (datame['brand'] == 'BRAND_J'),
    (datame['brand'] == 'BRAND_L'),
    (datame['brand'] == 'BRAND_P'),
    (datame['brand'] == 'BRAND_R'),
    (datame['brand'] == 'BRAND_S'),
    (datame['brand'] == 'BRAND_W'),
]
# create a list of the values we want to assign for each condition
values = [
    'Sari Wangi',
    'Segar Sari',
    'Teh Sisri',
    'Nutri Sari',
    'Kacang Dua Kelinci',
    'Lays',
    'Japota',
    'Milkita',
    'Super Zuper',
    'Jagoan Neon'
]
# create a new column and use np.select to assign values to it using our lists as arguments
datame['desc_brand'] = np.select(condition, values)
```

```python
datame
```

```python
ps.sqldf("""
select *,
    case when brand='BRAND_A' then 'Sari Wangi'
        when brand='BRAND_B' then 'Segar Sari'
        when brand='BRAND_C' then 'Teh Sisri'
        when brand='BRAND_H' then 'Nutri Sari'
        when brand='BRAND_J' then 'Kacang Dua Kelinci'
        when brand='BRAND_L' then 'Lays'
        when brand='BRAND_P' then 'Japota'
        when brand='BRAND_R' then 'Milkita'
        when brand='BRAND_S' then 'Super Zuper'
        when brand='BRAND_W' then 'Jagoan Neon'
    end as desc_brand_sql
from datame
""")
```

#### Tambahkan kolom baru bernama `cat_brand` & `cat_brand_sql` untuk mengkategorikan produk desc_brand dengan kriteria di bawah ini:

|desc_brand|cat_brand|
|---|---|
|Sari Wangi, Segar Sari, Teh Sisri, Nutri Sari|Minuman|
|Kacang Dua Kelinci, Japota, Lays|Makanan Ringan|
|Milkita, Super Zuper, Jagoan Neon|Permen|


```python
# Define Condition
condition = [
    (datame['desc_brand'] == 'Sari Wangi'),
    (datame['desc_brand'] == 'Segar Sari'),
    (datame['desc_brand'] == 'Teh Sisri'),
    (datame['desc_brand'] == 'Nutri Sari'),
    (datame['desc_brand'] == 'Kacang Dua Kelinci'),
    (datame['desc_brand'] == 'Japota'),
    (datame['desc_brand'] == 'Lays'),
    (datame['desc_brand'] == 'Milkita'),
    (datame['desc_brand'] == 'Super Zuper'),
    (datame['desc_brand'] == 'Jagoan Neon')
]
# create a list of the values we want to assign for each condition
values = [
    'Minuman',
    'Minuman',
    'Minuman',
    'Minuman',
    'Makanan Ringan',
    'Makanan Ringan',
    'Makanan Ringan',
    'Permen',
    'Permen',
    'Permen'
]
# create a new column and use np.select to assign values to it using our lists as arguments
datame['cat_brand'] = np.select(condition, values)
```

```python
datame
```

```python
ps.sqldf("""
select *,
    case when desc_brand='Sari Wangi' then 'Minuman'
        when desc_brand='Segar Sari' then 'Minuman'
        when desc_brand='Teh Sisri' then 'Minuman'
        when desc_brand='Nutri Sari' then 'Minuman'
        when desc_brand='Kacang Dua Kelinci' then 'Makanan Ringan'
        when desc_brand='Japota' then 'Makanan Ringan'
        when desc_brand='Lays' then 'Makanan Ringan'
        when desc_brand='Milkita' then 'Permen'
        when desc_brand='Super Zuper' then 'Permen'
        when desc_brand='Jagoan Neon' then 'Permen'
    end as cat_brand_sql
from datame
""")
```
