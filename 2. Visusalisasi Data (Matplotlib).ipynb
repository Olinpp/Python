#### <font color='black'>Set Properties</font>

Untuk memahami lebih lanjut bisa klik link berikut: [matplotlib setp](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.setp.html)

```python
#1. Import Package nya
import numpy as np
import matplotlib.pyplot as plt


#2. Prepare Datanya
x1 = np.array([1,3,5,7,9,11])
y1 = np.array([1,9,25,49,81,121])
x2 = np.array([3,4,16,18,20,22])
y2 = np.array([2,12,30,56,90,132]) 
    
    
    
#3. membuat plot 
plt1=plt.plot(x1,y1)
plt2=plt.plot(x2,y2)

#mengatur warna, garis, ketebalan dll
plt.setp(plt1, color='red', linestyle=':', linewidth=1, marker='o', markersize=6)
plt.setp(plt2, color='blue', linestyle=':', linewidth=0.5, marker='o', markersize=6)

#4. Tampilkan plot
plt.show()
```

### <font color=''>3. Working & plotting with Data</font>

Berikut ini tampilan data transaksi supermarket kita sepanjang tahun 2019, lengkap dengan masing-masing penjelasannya untuk tiap kolom.
* `order_id` : ID dari order/transaksi, 1 transaksi bisa terdiri dari beberapa produk, tetapi hanya dilakukan oleh 1 customer

* `city` : kota tempat toko terjadinya transaksi
* `province` : provinsi (berdasarkan city)
* `product_id` : ID dari suatu product yang dibeli
* `brand` : brand/merk dari product. Suatu product yang sama pasti memiliki brand yang sama
* `quantity` : Kuantitas/banyaknya product yang dibeli
* `item_price` : Harga dari 1 product (dalam Rupiah). Suatu product yang sama, bisa jadi memiliki harga yang berbeda saat dibeli


data:
https://storage.googleapis.com/dqlab-dataset/retail_raw_reduced.csv

```python
#load Dataset
dataset= pd.read_csv('https://storage.googleapis.com/dqlab-dataset/retail_raw_reduced.csv')
```

#### <font color='black'>Case Study 1</font>

Tampilkan GMV dari bulan ke bulannya. Dimana GMV didapat dari perkalian antara `item_price` & `quantity`

<details><summary>Click here for the Clue</summary>

* Buat kolom baru format bulan saja
* Buat kolom baru GMV
* Totalkan GMV berdasarkan bulan, buat dataframe baru
* Plot menggunakan line chart

</details>

```python
#penambahahan variable GMV
dataset['gmv']= dataset['item_price']*dataset['quantity']
dataset.head()
```

```python
#penambahan variable month
#karena kita hanya ingin menampilkan tahun dan bulan saja tanpa mencantumkan tanggal
dataset['order_month'] = dataset['order_date'].str.slice(0,7)
dataset.head()
#diambil dari order date diubah ke order month
```

```python
#buat dataframe baru untuk line chart
#monthly_agg_df= dataset.groupby('order_month',as_index='false')['gmv'].sum()

#bisa juga pake ini
monthly_agg_df= dataset.groupby('order_month')['gmv'].sum().reset_index()

#tampilkan data
monthly_agg_df
```

```python
#perbesar figure
plt.figure(figsize=(20,10))#boleh tidak dipakai

#membuat line plot
plot1=plt.plot(monthly_agg_df['order_month'],monthly_agg_df['gmv'])

#set xtick dan ytick untuk memperbagus tampilan
y_tick_label, location=plt.yticks()
plt.yticks(y_tick_label,(y_tick_label/1000000000).astype(int))
#dari angka miliar menjadi satuan rangenya

plt.xticks(rotation=45) #rotasi bagian x agar enak diliat

#set title & label
plt.title('GMV Value in 2019', color='darkblue', fontsize=17)
plt.xlabel('Order Month', fontsize=13, color='darkred')
plt.ylabel('Total GMV (Billion)', fontsize=13, color='darkred')

#set axis y agar start dari nol
plt.ylim(ymin=0)

#costume line
plt.setp(plot1,color='g', linestyle='-',linewidth=2, marker='o',)

#tambah grid garis bantu
plt.grid(color='grey', linestyle=':', linewidth=0.5)

#menambah komen di bagian tertentu seperti di grid 4
plt.text(2.5,5000000000,'Peningkatan GMV \n ada pada bulan 9 ke bulan 10')

#tampilkan plot
plt.show()
```

#### <font color='darkgreen'>C.3 Case Study 3</font>

Diasumsikan quantity dari transaksi per customer sangat mempengaruhi dengan besarnya GMV yang dihasilkan. Aapakah benar ?

<details><summary>Click here for the Clue</summary>

* Grup kan berdasarkan customer_id, hitung total gmv nya & total quantitynya
* Buat Dataframe nya
* Plot menggunakan scatter plot dan line plot untuk bantu polanya gunakan fungsi plot()

</details>

```python
# persiapan data
dataset_customer = dataset.groupby('customer_id').agg({'quantity':'sum','gmv':'sum'}).reset_index()
dataset_customer
```

```python
#figsize
plt.figure(figsize=(10,10))

#membuat plot
x = dataset_customer['quantity']
y = dataset_customer['gmv']

#tambahkan 'o' agar merubah garis mencati dot
plt.plot(x,y,'o', markersize=5)

#membuat line chart
#membuat garis regresi
m, b = np.polyfit(x,y,1)
plt.plot(x,m*x+b)

#set x & y ticks
y_tick_label, location=plt.yticks()
plt.yticks(y_tick_label,(y_tick_label/1000000).astype(int))
plt.xticks(rotation=35)

#set title & label
plt.title('Quantity vs GMV of Customer', color='black', fontsize=13)
plt.xlabel('Total Quantity', color='Black')
plt.ylabel('Total GMV in (Million)', color='Black')

#set starting print
plt.ylim(ymin=0, ymax=200000000)
plt.xlim(xmin=0, xmax=400)

#menghilangkan garis tepi
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#simpan gambar
plt.savefig('GMV')

#tampilkan plot
plt.show()
```

#### <font color='darkgreen'>C.4 Case Study 4</font>

Setelah melihat scatter plot tadi, terlihat seperti adanya outlier di GMV customer. Pastikan apakah benar ada outlier ? dan tangani outliernya tanpa menghapus data tersebut

<details><summary>Click here for the Clue</summary>

* Buat box plot menggunakan GMV
* Misal ada outlier, maka handle pakai cap dengan nilai maximal dan minimalnya. Asumsikan saja min nya 0, max nya= 36000000
* Cek kembali plot nya
</details>

```python
plt.figure(figsize=(25,10))

#membuat plot
plt.boxplot(dataset_customer['gmv'], vert=0)

#set x ticks & yticks
xtick_label, location = plt.xticks()
plt.xticks(xtick_label,(xtick_label/1000000).astype(int), fontsize=13)

#set title & label
plt.title('Box Plot of GMV (Before)', color='black')

#rescale
plt.xlim(xmin=0, xmax=500000000)

#tampilkan plot
plt.show()


```

```python
# Handling Outlier
# Asumsikan nilai maks nya adalah 36000000, sedangkan nilai min nya adalah 0
dataset_customer = ps.sqldf('''select customer_id,
quantity,
case when gmv > 36000000 then 36000000
when gmv < 0 then 0
else gmv
end as gmv
from dataset_customer limit 100
''')
dataset_customer

```

```python
#memastikan apakah masih ada outlier atau tidak
plt.figure(figsize=(25,10)) #ternyyata sudah tidak ada outlier

#membuat plot
plt.boxplot(dataset_customer['gmv'], vert=0)

#set x ticks & yticks
xtick_label, location = plt.xticks()
plt.xticks(xtick_label,(xtick_label/1000000).astype(int), fontsize=13)

#set title & label
plt.title('Box Plot of GMV (Before)', color='black')

#rescale
plt.xlim(xmin=0, xmax=38000000)

#save image
plt.savefig('outlier')

#tampilkan plot
plt.show()
```
